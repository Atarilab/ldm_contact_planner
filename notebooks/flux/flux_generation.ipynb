```python
import time
import torch
from diffusers import FluxPipeline
```

```python
pipe = FluxPipeline.from_pretrained("black-forest-labs/FLUX.1-dev", torch_dtype=torch.bfloat16)
pipe.enable_sequential_cpu_offload()
# pipe.enable_model_cpu_offload()
```

```python
i = 0
```

```python
prompt = "A photorealistic scene of a woman with a pony-tale picking one closed cardboard box. The box is placed on top of a white office table. The woman is standing with her feet on the floor and her arms extended to pick up the box. Her hands are firmly in contact with the box. She is wearing blue jeans a white t-shirt and red sneakers."
while i < 50:
    image = pipe(
        prompt,
        height=1024,
        width=1024,
        guidance_scale=3.5,
        num_inference_steps=20
    ).images[0]
    
    image.save(f"./generated_images/grasp/table/image{i}.png")

    i += 1
```

```python
prompt = "A photorealistic scene of a woman with a pony-tale picking one closed cardboard box from the floor. She is standing with her knees bent and arms extended. Her hands are firmly in contact with the box. She is wearing blue jeans a white t-shirt and red sneakers."
while i < 50:
    image = pipe(
        prompt,
        height=1024,
        width=1024,
        guidance_scale=3.5,
        num_inference_steps=20
    ).images[0]
    
    image.save(f"./generated_images/grasp/image{i}.png")

    i += 1
```

```python
prompt = "A realistic scene of a man pushing a platform truck. The man has brown hair and is wearing a white shirt, blue jeans, and red sneakers. His hands are grasping the platform truck handle. The platform truck has four wheels, is empty, has a large flat surface, and is grey."
while i < 50:
    start = time.time()
    image = pipe(
        prompt,
        height=1024,
        width=1024,
        guidance_scale=3.5,
        num_inference_steps=20
    ).images[0]
    print("Time to generate image: ", time.time() - start)
    
    image.save(f"./generated_images/push/image{i}.png")

    i += 1
```
